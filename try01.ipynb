```python

def sendCommand(commands):
    try:
        s = serial.Serial('COM7', 9600, timeout=1)

        # cmd = input("Enter command")
        # cmd = cmd + '\r'



        def command(cm):

            if int(cm) ==0:
                s.write(b'F')
                invar = s.readline()
                print(" INV " + str(invar))
                s.write(b'S')
                return "done"


            if int(cm)==1:
                s.write(b'G')
                invar = s.readline()
                print(" INV " + str(invar))
                s.write(b'S')
                return "done"


            if int(cm)==2:
                s.write(b'I')
                invar = s.readline()
                print(" INV " + str(invar))
                s.write(b'S')
                return "done"


            if int(cm)==3:
                s.write(b'L')
                invar = s.readline()
                print(" INV " + str(invar))
                s.write(b'S')
                return "done"


            if int(cm)==4:
                s.write(b'R')
                invar = s.readline()
                print(" INV " + str(invar))
                s.write(b'S')
                return "done"
            if int(cm)==5:
                s.write(b'H')
                invar = s.readline()
                print(" INV " + str(invar))
                s.write(b'S')
                return "done"


        data = commands.split()
        value=command(data[0])
        s.close()
        
        if value =="done":
            return "done"
        


    except serial.SerialException as e:
        # There is no new data from serial port
        print(str(e))
        return "error"

    except TypeError as e:
        print(str(e))
        s.port.close()
        sys.exit(1)
        return "error"

def predictionOFModel(image):
    start = time.time()
    
    probabilities = model.predict(image)
    # print(probabilities)
    # Print what the highest value probabilitie label
    print("output :",labels[np.argmax(probabilities)])
    
    mid = time.time()
    val=sendCommand(labels[np.argmax(probabilities)])
    
    
    end=time.time()
    
    print("prediction : ",mid-start)
    print("-----")
    print("commnd : ",end-mid)
    print("total : ",end-start)
    
    if val=="done":
        return "done",True
    else :
        return "error",False
    


os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'


# Load the model
model = load_model('keras_model.h5',compile=False)


# Grab the labels from the labels.txt file. This will be used later.
labels = open('labels.txt', 'r').readlines()

   

# CAMERA can be 0 or 1 based on default camera of your computer.
camUrl = "http://25.33.130.161:8080/shot.jpg"

def process():
#     # Grab the webcameras image.
#     ret, image = camera.read()
    imgResp = urllib.request.urlopen(camUrl)
    imgNp = np.array(bytearray(imgResp.read()),dtype=np.uint8)
    image = cv2.imdecode(imgNp,-1)
    if image is None:
        print("none")
        process()
    if image is not None :
        print(image.shape)
        origIMg = image
         # Show the image in a window
        origIMg=cv2.resize(origIMg,(500,500))
#         cv2.imshow('Webcam Image', origIMg)

        image = image[0:1080,250:1920]  #croping image
        # Resize the raw image into (224-height,224-width) pixels.
        image = cv2.resize(image, (224, 224), interpolation=cv2.INTER_AREA)
        cv2.imshow("CroppedImg",image)
       
        # Make the image a numpy array and reshape it to the models input shape.
        image = np.asarray(image, dtype=np.float32).reshape(1, 224, 224, 3)
        # Normalize the image array
        image = (image / 127.5) - 1
        # Have the model predict what the current image is. Model.predict
        # returns an array of percentages. Example:[0.2,0.8] meaning its 20% sure
        # it is the first label and 80% sure its the second label.
        value,call = predictionOFModel(image)
        if value=="done" and call==True:
            process()
        else:
        # print(np.argmax(probabilities))
            print("------eror-------\n")
            process()
        

        # Listen to the keyboard for presses.
#         keyboard_input = cv2.waitKey(1)
#     # 27 is the ASCII for the esc key on your keyboard.
#         if keyboard_input == 27:
#             break


if __name__=="__main__":
    process()
```
